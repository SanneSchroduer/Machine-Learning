## Step 1.2 Checking plausability & removing impossible values

```python
df.columns
```

### We are going to explore some variables to see if all values are plausable.
### We start with the variable weight, let's make a histogram and a boxplot.

```python
import seaborn as sns #this is the plotting library I'll be using 
import matplotlib.pyplot as plt

sns.distplot(df['weight'].dropna(), kde=False) #Selecting the rating column. I need to drop the NA's for the plot
plt.title('Weight per person')
plt.xlabel('Weigth (kg)')
plt.show()
```

```python
sns.boxplot(df['weight'].dropna()) #Selecting the rating column. I need to drop the NA's for the plot
plt.title('Weight per person')
plt.xlabel('Weigth (kg)')
plt.show()
```

### In these 2 plots, we see an outlier at around 700 kilograms, and 2 outliers at around 20 kilograms.
### We are going to remove this outliers from the dataset, because they are probably not correct.

```python
def clean_weight(x):
    if x > 200.0 or x < 40.0: 
        return float('NaN')
    else: 
        return x
df['weight'] = df['weight'].apply(clean_weight)
```

### Now, let's see what the histogram plot looks like without the outliers.
### We see that the outliers are gone and that the data now looks much more plausible.

```python
import seaborn as sns #this is the plotting library I'll be using 
import matplotlib.pyplot as plt

sns.distplot(df['weight'].dropna(), kde=False) #Selecting the rating column. I need to drop the NA's for the plot
plt.title('Weight per person')
plt.xlabel('Weigth (kg)')
plt.show()
```

# Step 3: Linear Regression

What is linear regression?
Linear regression can be used to model the relationship between a dependent and independent variable.
Using the value of the independent variable, you can use a linear model to predict the value of the independent variable.

## Step 3.1 Predictive model

Make dummy values for the categorical variables:

```python
dummies_gender = pd.get_dummies(df["gender"])
dummies_gender
```

```python
dummies_age = pd.get_dummies(df["age"])
dummies_age
```

```python
dummies_city = pd.get_dummies(df["city"])
dummies_city
```

```python
dummies_education = pd.get_dummies(df["education_1"])
dummies_education
```

```python
df = pd.concat([df, dummies_age, dummies_gender, dummies_city, dummies_education], axis=1) #add to dataframe
df.head()
```

```python
X = df[['Female', 'Male', '<25','25-34', '35-44', '45-54', '55-64', '65+']]
y = df['avg_steps']
```

```python
import math
from sklearn.linear_model import LinearRegression #remember, this is linear regression model from sk-learn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) #split the data, store it into different variables
X_train.head() #The train data
```

```python
lm = LinearRegression() #this creates a new LR model
lm.fit(X_train, y_train) #this "fits" the model on the training data
```

```python
y_test_p = lm.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=y_test_p)
plt.xlim(0, 12000) #This sets the x-axis limits to (0, 5e10 = 140000000000)
plt.ylim(0, 12000) #Ditto for y-axis. I want both axes to have the same length, so we can compare them
plt.plot([0, 12000], [0, 12000], color='red', lw=3) #This draws the straight red line, you can leave this out if you wish
plt.xlabel('Mean steps')
plt.ylabel('Mean steps (predicted)')
plt.show()
```

```python
r2_score(y_test, y_test_p)
```

```python
math.sqrt(mean_squared_error(y_test, y_test_p)) #this gives the model fit (R2) for the Y and predicted Y
```

## Step 3.2 Description of the model's results

* We made a linear regression model in which we tried to predict the mean number of steps based on the gender and the age of the participants.
* The R-squared is negative, indicating that less than 0% of the variance of the results could be explained by the model.
* The RMSE is 5693, meaning that the model is typically 5693 steps off in predicting the mean number of steps.
* Overall, the model performs very badly, and gender and age are not usefull in prediciting the mean number of steps.
